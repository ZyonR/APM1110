## Summative Assessment 1-Question 1

A company has three factories producing a product. Factory 1 produces \( x_1 \) of the product, factory 2 produces \( x_2 \), and factory 3 produces \( x_3 \), where their sum is equivalent to 1. The defective rates of the products are \( y_1 \), \( y_2 \), and \( y_3 \) respectively, where their sum is equivalent to 0.12. Write a program (user input for \( x_i \) and \( y_i \)) to calculate the probability that a randomly selected product is defective.

Note that:

$$
0.1 \leq x_i \leq 0.4 \text{ and } 0.01 \leq y_i \leq 0.05
$$

We shall be following the law of total probability which states that:

$$
\text{P(B)}=\sum_{i=1}^{n} \text{P(B}\cap\text{A}_i)=\sum_{i=1}^{n} \text{P(B}|\text{A}_i)\times\text{P(A}_i)
$$

### Input Proportions and defectives

```r
proportionProduced<-c()
defectives<-c()
clearConsole <- function() {
  cat("\033[2J")
}
validate<-function(numberInput,bools){
  if(!is.numeric(numberInput)){
    return(T)
  }
  if(numberInput<0){
    return(T)
  }
  if(bools){
    if(numberInput>0.4 |numberInput<0.1){
      return(T)
    }
  }else{
    if(numberInput>0.05 |numberInput<0.01){
      return(T)
    }
  }
  return(F)
}
insertNumber<-function(insertNum,bools){
  if (bools) {
    proportionProduced <<- c(proportionProduced, insertNum)
  } else {
    defectives <<- c(defectives, insertNum)
  }
}
ask_user<-function(tell_me){
  if(tell_me%%2==0){
    bools <- T
    inputNum <- as.numeric(readline(paste("Product Created By Factory",length(proportionProduced)+1,"(decimal, from 0.1 to 0.4): ")))
  }else{
    bools <- F
    inputNum <- as.numeric(readline(paste("Product Defectives Created By Factory",length(defectives)+1," (decimal, from 0.01 to 0.05): ")))
  }
  
  if(validate(inputNum,bools)){
      cat("Invalid Number Inserted!\n")
      clearConsole()
      return()
  }
  insertNumber(inputNum,bools)
}
while(length(defectives)!=2){
  ask_user(length(proportionProduced)+length(defectives))
}
proportionProduced <- c(proportionProduced, 1-sum(proportionProduced))
defectives <- c(defectives, 0.12-sum(defectives))
print(paste("Proportion: ",proportionProduced))
print(paste("Defectives: ",defectives))
print(paste("Probability of product being defective: ",sum(proportionProduced*defectives)))
```
